```python

data_dir = '../New folder/Newtonian Fluid/' 
file_names = [
    'NUS_Nu J=4 (P=1N).dat',
    'NUS_Nu J=6 (P=1N).dat',
    'NUS_Nu J=8 (P=1N).dat',
    'NUS_Nu Kr = 0.4 (P=1N).dat',
    'NUS_Nu Kr = 1.4 (P=1N).dat',
    'NUS_Nu kr = 2.4 (P=1N).dat',
    'NUS_Nu  m = 0.1 (P=1N).dat',
    'NUS_Nu  m = 0.3 (P=1N).dat',
    'NUS_Nu m = 0.5 (P=1N).dat',
    'NUS_Nu M = 1 (P = 1N).dat',
    'NUS_Nu M = 3 (P = 1N).dat',
    'NUS_Nu M = 5 (P = 1N).dat',
    'NUS_Nu Nr = 1 (P=1N).dat',
    'NUS_Nu Nr = 2 (P=1N).dat',
    'NUS_Nu Nr = 3 (P=1N).dat'
]

dataframes = []
for file_name in file_names:
    file_path = os.path.join(data_dir, file_name)
    if os.path.exists(file_path):
        df = pd.read_csv(file_path, delimiter='\t', header=None)  
        dataframes.append(df)
        print(f" {file_path}")
    else:
        print(f"{file_path}")
if dataframes:
    combined_data = pd.concat(dataframes, ignore_index=True)
    print(combined_data.head())
else:
    print("No data files were read.")
```
