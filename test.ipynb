```python
file_path = "nus.csv"  
df = pd.read_csv(file_path, delim_whitespace=True, header=None)
df.columns = ['X', 'Y']
print(df.head())
```

```python
plt.scatter(df['X'], df['Y'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Newtonian')
```

```python
import seaborn as sns
cols = ["#6daa9f","#774571"]
sns.countplot(x=df["X"], palette=cols)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define palette colors
cols = ["#C2C4E2", "#EED4E5"]

# Create a scatter plot
sns.scatterplot(x=df["X"], y=df["Y"], palette=cols)

# Show the plot
plt.show()
```

```python

plt.figure(figsize=(12, 8))
# Create a line plot
Time_series = sns.lineplot(x="X", y="Y", data=df, color="#C2C4E2")
# Set title and labels
Time_series.set_title("Newtonian")
Time_series.set_ylabel("Y")

# Show the plot
plt.show()
```

```python
colours = ["#D0DBEE", "#C2C4E2", "#EED4E5", "#D1E6DC", "#BDE2E2"]
plt.figure(figsize=(12,8))
Days_of_week=sns.barplot(x="X", y="Y", data=df, ci =None,palette = colours)
Days_of_week.set_title("Newtonian")
Days_of_week.set_ylabel("Y")
Days_of_week.set_xlabel("X")
```

```python
colours = ["#D0DBEE", "#6daa9f","#774571", "#D1E6DC", "#BDE2E2"]
plt.figure(figsize=(20,10))
sns.boxenplot(data =df,palette = colours)
plt.xticks(rotation=90)
plt.show()
```
